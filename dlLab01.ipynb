```python
import matplotlib.pyplot as plt
```

```python
sample2_uniform = np.random.rand(100000, 1)
```

```python
sample2_normal = np.random.normal(size=(100000, 1))
```

```python
bins = 50
plt.hist(sample2, density=True, bins=bins, histtype="step", color="blue", label="Exponential")
plt.hist(sample2_uniform, density=True, bins=bins, histtype="step", color="green", label="Uniform")
plt.hist(sample2_normal, density=True, bins=bins, histtype="step", color="red", label="Normal")

plt.axis([0, 10, 0, 1.1])

plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend(loc="upper right")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = X**2 + Y**2')
ax.set_title('3D Plot of Z = X**2 + Y**2')

plt.show()
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
```

```python
df.head()
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

pearson_corr = data.corr(method='pearson')

spearman_corr = data.corr(method='spearman')

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={'size': 10}, ax=axes[0])
axes[0].set_title('Pearson Correlation')

sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={'size': 10}, ax=axes[1])
axes[1].set_title('Spearman Correlation')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')
```

```python
desired_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
filtered_df = df[desired_columns]
```

```python
print(filtered_df.head(20))
```

```python
mean_values = filtered_df.mean()
```

```python
print("Mean values:")
print(mean_values)
```

```python
std_values = filtered_df.std()
```

```python
print("\nStandard deviation values:")
print(std_values)
```

```python
quartile_values = filtered_df.quantile([0.25, 0.5, 0.75])
```

```python
print("\nQuartile values:")
print(quartile_values)
```

```python
plt.figure(figsize=(15, 4))

for i, column in enumerate(filtered_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(filtered_df[column], bins=20, edgecolor='black', alpha=0.7)
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(filtered_df)
plt.show()
```

```python
g = sns.PairGrid(filtered_df)
g.map_diag(sns.histplot, kde=True)
g.map_upper(sns.histplot, bins=20, cbar=True)
g.map_lower(sns.kdeplot, cmap='Blues_d')
g.fig.suptitle("Pair Plot of Features", size=16)
plt.subplots_adjust(top=0.9)
plt.show()
```
